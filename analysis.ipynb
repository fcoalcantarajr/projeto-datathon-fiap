```python
import pandas as pd

# Read the JSON file
df_prospects = pd.read_json('data/prospects.json', orient='index')

# Ensure 'prospects' column exists for all rows
df_prospects['prospects'] = df_prospects.get('prospects', pd.Series([[]]*len(df_prospects), index=df_prospects.index))

# Explode the 'prospects' list so each prospect gets its own row
df_exploded = df_prospects.explode('prospects')

# Drop rows where 'prospects' is still empty (i.e., NaN after explode)
df_exploded = df_exploded.dropna(subset=['prospects'])

# Normalize the 'prospects' dictionaries into columns
if not df_exploded.empty:
    df_normalized = pd.json_normalize(df_exploded['prospects'])
    # Combine with the index (IDs)
    df_result = df_exploded.drop(columns=['prospects']).join(df_normalized)
else:
    df_result = pd.DataFrame()  # No prospects to normalize

df_result.head()
```

```python
df_prospects = df_result
```
