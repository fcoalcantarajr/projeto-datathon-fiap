```python
# Read the JSON file
df_prospects = pd.read_json('data/prospects.json', orient='index')

# Ensure 'prospects' column exists for all rows
df_prospects['prospects'] = df_prospects.get('prospects', pd.Series([[]]*len(df_prospects), index=df_prospects.index))

# Explode the 'prospects' list so each prospect gets its own row
df_exploded = df_prospects.explode('prospects')

# Drop rows where 'prospects' is still empty (i.e., NaN after explode)
df_exploded = df_exploded.dropna(subset=['prospects'])

# Normalize the 'prospects' dictionaries into columns
if not df_exploded.empty:
    df_normalized = pd.json_normalize(df_exploded['prospects'])
    # Combine with the index (IDs)
    df_result = df_exploded.drop(columns=['prospects']).join(df_normalized)
else:
    df_result = pd.DataFrame()  # No prospects to normalize

df_result.head()
```

##### Usar df_applicants_filtered e df_vagas_filtered para fazer o match entre candidatos e vagas

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download necessary NLTK data
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize stop words for Portuguese and lemmatizer
stop_words = set(stopwords.words('portuguese'))
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')

# Function to clean text: remove non-alphanumeric characters, lowercase, lemmatize, and remove stop words
def clean_text(text):
    if isinstance(text, str):
        # Remove non-alphanumeric characters
        text = re.sub(r'[^\w\s]', '', text, re.UNICODE)
        # Lowercase the text
        text = text.lower()
        # Split the text into words, lemmatize each word, and join them back into a string
        text = [lemmatizer.lemmatize(token) for token in text.split(" ")]
        text = [lemmatizer.lemmatize(token, "v") for token in text]
        # Remove stop words
        text = [word for word in text if not word in stop_words]
        # Join the cleaned words
        text = " ".join(text)
        return text
    else:
        return ''

# Apply the cleaning function to all object columns in df_applicants_filtered
for col in df_applicants_filtered.columns:
    if df_applicants_filtered[col].dtype == 'object':
        df_applicants_filtered[col] = df_applicants_filtered[col].astype(str).apply(clean_text)

# Apply the cleaning function to all object columns in df_vagas_filtered
for col in df_vagas_filtered.columns:
    if df_vagas_filtered[col].dtype == 'object':
        df_vagas_filtered[col] = df_vagas_filtered[col].astype(str).apply(clean_text)

# Print the head of the cleaned dataframes
print("Cleaned df_applicants_filtered:")
print(df_applicants_filtered.head())
print("\nCleaned df_vagas_filtered:")
print(df_vagas_filtered.head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Dictionary to store TF-IDF matrices for applicants
tfidf_applicants = {}

# Iterate through each column of df_applicants_filtered
for col in df_applicants_filtered.columns:
    # Fit and transform the text data
    tfidf_applicants[col] = tfidf_vectorizer.fit_transform(df_applicants_filtered[col])

# Dictionary to store TF-IDF matrices for vagas
tfidf_vagas = {}

# Iterate through each column of df_vagas_filtered
for col in df_vagas_filtered.columns:
    # Fit and transform the text data
    tfidf_vagas[col] = tfidf_vectorizer.fit_transform(df_vagas_filtered[col])

print("TF-IDF matrices for applicants:")
print(tfidf_applicants.keys())
print("\nTF-IDF matrices for vagas:")
print(tfidf_vagas.keys())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# Função para calcular a similaridade entre uma vaga e todos os candidatos
def rank_applicants_for_job(vaga_idx, df_applicants_filtered):
    # Colunas para comparar (ajuste conforme necessário)
    columns_to_compare = [
        ('titulo_vaga', 'titulo_profissional'),
        ('cidade', 'endereco'),
        ('nivel profissional', 'nivel_profissional'),
        ('nivel_ingles', 'nivel_ingles'),
        ('nivel_espanhol', 'nivel_espanhol'),
        ('areas_atuacao', 'area_atuacao'),
        ('principais_atividades', 'conhecimentos_tecnicos'),
        ('competencia_tecnicas_e_comportamentais', 'certificacoes'),
        # 'demais_observacoes' e 'habilidades_comportamentais_necessarias' podem ser incluídas se fizer sentido
    ]
    
    total_score = np.zeros(len(df_applicants_filtered))
    for vaga_col, applicant_col in columns_to_compare:
        if vaga_col in df_vagas_filtered.columns and applicant_col in df_applicants_filtered.columns:
            # Combine texts for fitting vectorizer
            vaga_text = df_vagas_filtered.loc[vaga_idx, vaga_col]
            applicants_text = df_applicants_filtered[applicant_col].fillna("").astype(str).tolist()
            combined_texts = [vaga_text] + applicants_text
            vectorizer = TfidfVectorizer()
            tfidf_matrix = vectorizer.fit_transform(combined_texts)
            vaga_vec = tfidf_matrix[0]
            applicants_mat = tfidf_matrix[1:]
            # Calcula similaridade entre a vaga e todos os candidatos para esta coluna
            sim = cosine_similarity(vaga_vec, applicants_mat).flatten()
            total_score += sim
    
    # Retorna os índices dos top 10 candidatos
    top10_idx = np.argsort(-total_score)[:10]
    return df_applicants_filtered.iloc[top10_idx].assign(score=total_score[top10_idx])

# Função para mostrar os top 10 candidatos para uma vaga selecionada pelo índice
def mostrar_top10_para_vaga(vaga_idx, df_vagas_filtered, df_applicants_filtered):
    top10 = rank_applicants_for_job(vaga_idx, df_applicants_filtered)
    print(f"Top 10 candidatos para a vaga: {df_vagas_filtered.loc[vaga_idx, 'titulo_vaga']}")
    display(top10[['nome', 'titulo_profissional', 'area_atuacao', 'score']])

# Demonstração com a primeira vaga (índice 0)
mostrar_top10_para_vaga(0, df_vagas_filtered, df_applicants_filtered)
```
